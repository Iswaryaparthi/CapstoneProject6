```python
df5 = pd.read_csv('RegistrationdatefromJan2017onwards.csv')
df5.head()
```

#### convert the year and month as numeric value in remaining_lease column

```python
df5['remaining_lease'] = df5['remaining_lease'].str.split(' ').str[0].astype('int64')
```

```python
df5.head()
```

#### merging the five datasets

```python
df_combined = pd.concat([df1,df2,df3,df4,df5], ignore_index=True)
df_combined.head()
```

```python
df_combined.isnull().sum()
```

```python
df_combined[['year', 'month']] = df_combined['month'].str.split('-',expand=True)
```

```python
df_combined.head()
```

```python
df_combined['year'] = pd.to_numeric(df_combined['year'])
df_combined['month'] = pd.to_numeric(df_combined['month'])
```

```python
df_combined.head()
```

#### fill the null values in remaining_lease column

```python
df_combined['remaining_lease'].fillna(99-(df_combined['year']-df_combined['lease_commence_date']),inplace=True)
df_combined.head()
```

```python
df_combined.isnull().sum()
```

```python
df_combined.to_csv('resale.csv', index=False)
```
